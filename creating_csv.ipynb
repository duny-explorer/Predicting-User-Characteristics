```python
import pandas as pd
```

```python
df = pd.read_csv('csv_data/merged_df.csv')
```

```python
df.info()
```

```python
def calculate_median_percent(df):

    df = df[df['duration'] > 0]
    df['percent_watched'] = df['total_watchtime'] / df['duration'] * 100


    median_percent_per_category = df.groupby('category')['percent_watched'].median().reset_index()

    median_percent_per_category.columns = ['category', 'median_percent_watched']

    return median_percent_per_category


median_percent_df = calculate_median_percent(df)
```

```python
median_percent_df['median_percent_watched'] = median_percent_df['median_percent_watched'] * 100
```

```python
import math
```

```python
median_percent_df['median_percent_watched'] = median_percent_df['median_percentwatched'].apply(lambda x: math.ceil(x))
```

```python
median_percent_df
```

```python
median_percent_df.to_csv('median_percent_watched_category.csv')
```

```python
def calculate_median_age_category(df):

    df = df[df['duration'] > 0]
    df['percent_watched'] = df['total_watchtime'] / df['duration'] * 100


    median_percent_per_category = df.groupby('age')['percent_watched'].median().reset_index()

    median_percent_per_category.columns = ['age', 'median_percent_watched']

    return median_percent_per_category


median_percent_df = calculate_median_percent_age(df)
```
